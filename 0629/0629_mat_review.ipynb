```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
x=range(0,100)
y=[v*v for v in x]
ax1.plot(x,y)
ax2.bar(x,y)
plt.show()
```

```python
x=range(0,100)
y=[v*v for v in x]

fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)
ax1.plot(x,y)

ax2 = fig.add_subplot(2,1,2)
ax2.bar(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0.0, 2 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
ax1.plot(x, sin_y, 'b--')
ax2.plot(x, cos_y, 'r--')
plt.show()
```

```python
x = np.arange(10)
plt.plot(x, 1*x, label='y = %ix' % 1)
plt.plot(x, 2*x, label='y = %ix' % 2)
plt.legend()
plt.show()
```

```python
for i in range(5):
 plt.plot(x, i*x, label='y = %ix' % i)
plt.legend()
plt.show()
```
