### 공공데이터를 사용하여 우리나라 지도에 시각화 수행하기

 - url: https://www.data.go.kr/data/15043191/fileData.do
 - 국가보훈처 보훈의료 위탁병원 코로나-19 병원 정보: 보훈대상자 의료 제공 위탁병원의 코로나-19 관련 병원 운영 현황 정보
(코로나 관련 국민안심병원, 감염병전담기관, 감염병전담기관 외래진료 정보)

```python
import pandas as pd
import numpy as np
import folium
from folium import plugins
```

1. 주소 값을 구글스프레드시트(Geocode by Awesome Table)에서 위도,경도 값 변환하여 파일로 저장

```python
dat = pd.read_csv("./corona_hospital.csv" , engine ='python', encoding='EUCKR')
```

2. 결측치 확인 후 가공 

```python
dat.info()
dat.isnull().sum()
```

```python
dat=dat.dropna(axis=0, subset=["Latitude", "Longitude"])
```

```python
dat.info()
```

3. 데이터 인덱싱: 필요정보 가져오기

```python
dat = dat.loc[:, ["의료기관명", "Latitude", "Longitude"]]
```

```python
dat_map = folium.Map(location=[37.5665, 126.9780])
dat_map
```

4. numpy자료로 만들기

```python
name_of_hospital = dat.loc[:,'의료기관명']
name_of_hospital = name_of_hospital.to_numpy()
type(name_of_hospital)
```

```python
temp = np.array(zip(dat["Latitude"], dat["Longitude"]))
```

```python
geocodes = np.array(list(zip(dat["Latitude"], dat["Longitude"])))
geocodes[0:10]
```

5. Marker를 군집화 시키기 - MarkerCluster

```python
plugins.MarkerCluster(geocodes, popups=name_of_hospital).add_to(dat_map)
```

```python
dat_map
```

```python
dat_map.save('bohun_corona_hospital.html')
```
