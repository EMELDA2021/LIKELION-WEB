## 데이터 정제하기

2021년 1월 1일 ~ 2021년 6월 15일의 데이터만 가져오기

```python
covid = covid[covid["date"] >= "2021-01-01"]
covid = covid[covid["date"] <= "2021-06-15"]
```

필요한 columns만 가져오기

```python
covid = covid[["continent", "location", "date", "total_cases", "new_cases_smoothed", "total_deaths", "new_deaths_smoothed", "population"]]
covid.info()
```

인구 정보가 없는 국가와 대륙별 집계 데이터를 제거하기

```python
covid = covid.dropna(axis=0, subset=["continent", "population"])
covid.info()
```

## 필요한 정보 추가하기

인구 대비 하루 확진자수와 하루 사망자수를 추가하기

```python
covid["death_per_population"] = covid["new_deaths_smoothed"] / covid["population"] * 100
covid["cases_per_population"] = covid["new_cases_smoothed"] / covid["population"] * 100
covid
```

### 대륙별 하루 확진자수 그래프 그리기

```python
plt.figure(figsize=(12, 10))
sns.lineplot(x="date", y="new_cases_smoothed", data=covid[covid["continent"] == "Asia"], ci=None, label="Asia")
sns.lineplot(x="date", y="new_cases_smoothed", data=covid[covid["continent"] == "Africa"], ci=None, label="Africa")
sns.lineplot(x="date", y="new_cases_smoothed", data=covid[covid["continent"] == "North America"], ci=None, label="N. America")
sns.lineplot(x="date", y="new_cases_smoothed", data=covid[covid["continent"] == "South America"], ci=None, label="S. America")
sns.lineplot(x="date", y="new_cases_smoothed", data=covid[covid["continent"] == "Europe"], ci=None, label="Europe")
sns.lineplot(x="date", y="new_cases_smoothed", data=covid[covid["continent"] == "Oceania"], ci=None, label="Oceania")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

### 인구 대비 대륙별 하루 확진자수 그래프 그리기

```python
plt.figure(figsize=(12, 10))
sns.lineplot(x="date", y="cases_per_population", data=covid[covid["continent"] == "Asia"], ci=None, label="Asia")
sns.lineplot(x="date", y="cases_per_population", data=covid[covid["continent"] == "Africa"], ci=None, label="Africa")
sns.lineplot(x="date", y="cases_per_population", data=covid[covid["continent"] == "North America"], ci=None, label="N. America")
sns.lineplot(x="date", y="cases_per_population", data=covid[covid["continent"] == "South America"], ci=None, label="S. America")
sns.lineplot(x="date", y="cases_per_population", data=covid[covid["continent"] == "Europe"], ci=None, label="Europe")
sns.lineplot(x="date", y="cases_per_population", data=covid[covid["continent"] == "Oceania"], ci=None, label="Oceania")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
