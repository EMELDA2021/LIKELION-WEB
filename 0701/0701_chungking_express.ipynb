```python
import nltk
from konlpy.tag import Kkma ### 꼬꼬마
from konlpy.tag import Hannanum ### 한나눔
```

```python
### wordcloud와 이미지 표시
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

```python
dat_ko = open("chungking_express.csv").read()
dat_ko[1:1000]
```

```python
from konlpy.tag import Okt
t=Okt()
```

```python
doc_nouns1=t.nouns(dat_ko)
print(len(doc_nouns1))
doc_nouns1[0:10]
```

```python
ko=nltk.Text(doc_nouns1,name='중경삼림')
print(type(ko))
print(len(ko.tokens))
```

```python
ko.vocab().most_common(20)
```

```python
most_fre=ko.vocab().most_common(50)
most_fre
```

```python
stop_words = ['홍콩','영화','홍콩영화','내','더','난','접','이자','김','면',
              '볼','만하','드','리밍','이영화','등','곳','요',
              '걸','때','그냥']
new_ko = [ ]
for one_word in ko:
  if one_word not in stop_words:
    new_ko.append(one_word)
```

```python
### nltk Text 객체 만들기
new_ko = nltk.Text(new_ko, name="중경삼림")
plt.figure(figsize=(12,6))
new_ko.plot(50)
```

```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
```

```python
chunking_mask = np.array(Image.open("can.jpg"))
```

```python
data = new_ko.vocab().most_common(1000)
```

```python
wc = WordCloud(background_color='white', 
 max_words=500,
 mask=chunking_mask, 
 contour_width=3, 
 contour_color="steelblue", 
 font_path=path).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(10,16))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
