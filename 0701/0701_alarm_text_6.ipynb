```python
page_num=driver.find_element_by_xpath('//*[@id="maxPage"]').text
print("페이수는?",page_num)
```

```python
all_date=[]
all_time=[]
all_area=[]
count=0
while count < (int(num)+1):        
    for i in range(1,(int(page_num)+1)):
        page_input=driver.find_element_by_xpath('//*[@id="bbs_page"]')
        page_input.clear()
        page_input.send_keys(f'{i}')
        page_click=driver.find_element_by_xpath('//*[@id="content"]/div[4]/div/a[3]')
        page_click.click()
        time.sleep(3)
        
        for i in range(0,10):    
            count+=1
            if count > int(num):
                break
            one_list=driver.find_element_by_xpath('//*[@id="bbs_tr"]')
            tmp=f'//*[@id="bbs_tr_{i}_bbs_title"]'
        
            sel_num=driver.find_element_by_xpath(tmp).text.split(' ')
            all_date.append(sel_num[0])
            all_time.append(sel_num[1])
            all_area.append(sel_num[2].strip('재난문자'))
    time.sleep(3)
    try:    
        pass
    except NoSuchElementException or StaleElementReferenceException:
        pass
    
print(count)
print(len(all_date))
print(len(all_time))
print(len(all_area))
print(all_date)
```
