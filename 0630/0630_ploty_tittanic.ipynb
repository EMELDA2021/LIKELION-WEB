```python
import plotly
import cufflinks as cf
import pandas as pd
import numpy as np
import sys
import seaborn as sns
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python
Age_mean=titanic['age'].mean()
titanic['age']=titanic['age'].fillna(Age_mean)
titanic['age'].isnull().sum()
```

```python
titanic['embarked']=titanic['embarked'].fillna('S')
titanic['embarked'].isnull().sum()
```

```python
titanic['deck'].unique()
```

```python
titanic.deck.value_counts()
```

```python
titanic['deck']=titanic['deck'].fillna('C')
titanic['deck'].isnull().sum()
```

```python
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
titanic.adult_male.head()
```

Q.성별로 비교할 때 생존율은?

```python
titanic.iplot(kind='bar', x='survived', 
              y=['sex','adult_male'], 
              xTitle='survived', yTitle = 'sex&adult_male')
```

결론:여성 생존이 높고, 남성 생존이 낮은 이유는 성인남자는 사망이 더 높기 때문이다.

```python
titanic.survived.astype(str)
```

Q.동행인이 있다면 생존율이 높을까?

```python
titanic.iplot(kind='bar', x='survived', 
              y=['sibsp', 'parch' ], barmode='stack',
              xTitle='survived', yTitle = 'survived per sibsp&parch')
```

결론: 형제자매가 있는 경우 부모,자녀가 있는 사람보다 생존율이 높다

```python
titanic.corr()
```

생존의 영향을 미치는 인자 찾기(상관관계)

```python
sel=['survived','pclass','age','sibsp','parch','fare']
titanic[sel].iplot(kind='bar',x='survived')
```

Q.승선항에 따른 생존율?

```python
titanic[['survived','embarked']].iplot(kind='bar',x='embarked',y='survived')
```

```python
titanic.embarked.value_counts()
```

결론: s항에서 탑승시 생존율이 높지만, s항에 인원자체가 높아 높은값으로 나옴

Q.나이와 클래스의 편차를 보자.
