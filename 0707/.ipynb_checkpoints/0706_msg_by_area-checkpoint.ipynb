```python
import pandas as pd
from datetime import datetime
import os
```

```python
msg_total_id=pd.read_csv('./msg_total_id.csv',engine ='python')
```

```python
msg_total_id['Area']=msg_total_id['Area'].fillna('권역')
```

```python
msg_total_id.isnull().sum()
```

```python
msg_total_id['create_date']=pd.to_datetime(msg_total_id['create_date'])
```

```python
msg_total_id
```

### 지역별 검사장려 문자 나누기

```python
plz_be_tested = msg_total_id[msg_total_id["msg"].str.contains("선별진료소") |\
                          msg_total_id["msg"].str.contains("무기명") |\
                          msg_total_id["msg"].str.contains("증상") |\
                          msg_total_id["msg"].str.contains("발현") |\
                          msg_total_id["msg"].str.contains("의심") |\
                          msg_total_id["msg"].str.contains("가까운") |\
                          msg_total_id["msg"].str.contains("발열") |\
                          msg_total_id["msg"].str.contains("오한") |\
                          msg_total_id["msg"].str.contains("기침") |\
                          msg_total_id["msg"].str.contains("두통") |\
                          msg_total_id["msg"].str.contains("구토") |\
                          msg_total_id["msg"].str.contains("설사") |\
                          msg_total_id["msg"].str.contains("1339")
                         ]
```

```python
plz_be_tested=plz_be_tested.drop(['location_id','location_name','md101_sn','msg','send_platform','location_sido'],axis=1)
plz_be_tested=plz_be_tested.reset_index(drop=True)
plz_be_tested
```

```python
plz_be_tested['Area'].unique()
```

```python
plz_be_tested['Area'].value_counts()
```

```python
a=pd.DataFrame(plz_be_tested.groupby(["Area", plz_be_tested.create_date.dt.date]).size()).loc["수도권", :]
b=pd.DataFrame(plz_be_tested.groupby(["Area", plz_be_tested.create_date.dt.date]).size()).loc["충청권", :]
c=pd.DataFrame(plz_be_tested.groupby(["Area", plz_be_tested.create_date.dt.date]).size()).loc["호남권", :]
d=pd.DataFrame(plz_be_tested.groupby(["Area", plz_be_tested.create_date.dt.date]).size()).loc["경북권", :]
e=pd.DataFrame(plz_be_tested.groupby(["Area", plz_be_tested.create_date.dt.date]).size()).loc["경남권", :]
f=pd.DataFrame(plz_be_tested.groupby(["Area", plz_be_tested.create_date.dt.date]).size()).loc["강원권", :]
g=pd.DataFrame(plz_be_tested.groupby(["Area", plz_be_tested.create_date.dt.date]).size()).loc["제주", :]
h=pd.DataFrame(plz_be_tested.groupby(["Area", plz_be_tested.create_date.dt.date]).size()).loc["권역", :]
```

```python
plz_area_msg=pd.concat([a,b,c,d,e,f,g,h],ignore_index = True, axis=1)
plz_area_msg.columns=['수도권','충청권','호남권','경북권','경남권','강원권','제주','권역']
```

```python
plz_area_msg=plz_area_msg.fillna(0)
plz_area_msg.sort_index()
```

```python
plz_area_msg1=plz_area_msg.copy()
plz_area_msg1.loc['all_count']=[plz_area_msg['수도권'].sum(),plz_area_msg['충청권'].sum(),
                                plz_area_msg['호남권'].sum(),plz_area_msg['경북권'].sum(),
                               plz_area_msg['경남권'].sum(),plz_area_msg['강원권'].sum(),
                               plz_area_msg['제주'].sum(),plz_area_msg['권역'].sum(),]
plz_area_msg1
```

### 검사독려문자 평활화

```python
plz_area_msg_smoothed = plz_area_msg.rolling(7, min_periods=1).mean()
plz_area_msg_smoothed
```

### 지역별 전체문자 나누기

```python
total_msg=msg_total_id.drop(['location_id','location_name','md101_sn','msg','send_platform','location_sido'],axis=1)
total_msg=total_msg.reset_index(drop=True)
total_msg.head()
```

```python
total_msg['Area'].unique()
```

```python
total_msg['Area'].value_counts()
```

```python
a=pd.DataFrame(total_msg.groupby(["Area", total_msg.create_date.dt.date]).size()).loc["수도권", :]
b=pd.DataFrame(total_msg.groupby(["Area", total_msg.create_date.dt.date]).size()).loc["충청권", :]
c=pd.DataFrame(total_msg.groupby(["Area", total_msg.create_date.dt.date]).size()).loc["호남권", :]
d=pd.DataFrame(total_msg.groupby(["Area", total_msg.create_date.dt.date]).size()).loc["경북권", :]
e=pd.DataFrame(total_msg.groupby(["Area", total_msg.create_date.dt.date]).size()).loc["경남권", :]
f=pd.DataFrame(total_msg.groupby(["Area", total_msg.create_date.dt.date]).size()).loc["강원권", :]
g=pd.DataFrame(total_msg.groupby(["Area", total_msg.create_date.dt.date]).size()).loc["제주", :]
h=pd.DataFrame(total_msg.groupby(["Area", total_msg.create_date.dt.date]).size()).loc["권역", :]
```

```python
total_area_msg=pd.concat([a,b,c,d,e,f,g,h],ignore_index = True, axis=1)
total_area_msg.columns=['수도권','충청권','호남권','경북권','경남권','강원권','제주','권역']
```

```python
total_area_msg=total_area_msg.fillna(0)
total_area_msg.sort_index()
```

```python
total_area_msg1=total_area_msg.copy()
total_area_msg1.loc['all_count']=[total_area_msg['수도권'].sum(),total_area_msg['충청권'].sum(),
                                total_area_msg['호남권'].sum(),total_area_msg['경북권'].sum(),
                               total_area_msg['경남권'].sum(),total_area_msg['강원권'].sum(),
                               total_area_msg['제주'].sum(),total_area_msg['권역'].sum(),]
total_area_msg1
```

### 전체문자 평활화

```python
total_area_msg_smoothed = total_area_msg.rolling(7, min_periods=1).mean()
total_area_msg_smoothed
```
