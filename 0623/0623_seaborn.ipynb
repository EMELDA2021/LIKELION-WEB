```python
import seaborn as sns
sns.set_theme(style="whitegrid")

# Load the example dataset for Anscombe's quartet
df = sns.load_dataset("tips")

# Show the results of a linear regression within each dataset
sns.lmplot(x="tip", y="total_bill", col="sex", hue="day", data=df,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
```

<앤스콤 4분할 그래프><data='tips>

implot>데이터를 scatter처럼 그리고 직선으로 regression한 그림도 같이 그려주고 유효범위도 ci로 잡아줌
hue옵션으로 구분할 수 있고, palette 옵션으로 색상지정 가능함

whitegrid>바탕그래프(실선)
col_wrap>열갯수
col:범주형 변수
hue:범주형 변수,그래프별 색상
ci=None:회귀추정의 신뢰구간(1~100)
scatter_kws:닷 관련 파라미터
line_kws:라인 관련 파라미터

결론
1.남성이 여성보다 보다 많은 팁과 구매를 하였다.
2.요일로는 토요일이 공통적으로 많은 팁을 보였다.
3.남성,일요일의 경우 회귀선과 특이치와 차이가 크게 나타났다.

```python
fig = plt.figure(figsize=(10,6))
fig.patch.set_facecolor('xkcd:white')

axis1 = fig.add_subplot(2, 2, 1)
axis1 = sns.barplot(x='species', y='sepal_length', data=iris)
axis1.plot()

axis2 = fig.add_subplot(2, 2, 2)
axis2 = sns.barplot(x='species', y='sepal_width', data=iris)
axis2.plot()

axis3 = fig.add_subplot(2, 2, 3)
axis3 = sns.barplot(x='species', y='petal_length', data=iris)
axis3.plot()

axis4 = fig.add_subplot(2, 2, 4)
axis4 = sns.barplot(x='species', y='petal_length', data=iris)
axis4.plot()
```
