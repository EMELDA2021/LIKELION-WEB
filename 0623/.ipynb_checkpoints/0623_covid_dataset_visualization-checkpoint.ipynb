```python
import pandas as pd
import os
!git clone https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-data.csv
```

```python
covid_data = pd.read_csv("./owid-covid-data.csv")
```

```python
covid_data.info()
```

```python
covid_data_after_2021 = covid_data[covid_data["date"] >= "2021-01-01"]
covid_data_after_2021 = covid_data_after_2021[covid_data_after_2021["date"] <= "2021-06-15"]
```

```python
covid_data_filtered = covid_data_after_2021[["continent", "location", "date", "total_cases", "new_cases", "total_deaths", "new_deaths_smoothed" ,"hosp_patients","positive_rate","new_vaccinations_smoothed","population","reproduction_rate"]]
```

```python
covid_data_filtered.info()
```

```python
covid_data_filtered.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 10))
sns.lineplot(x="date", y="hosp_patients", data=covid_data_filtered[covid_data_filtered["continent"] == "Asia"], label='asia')
sns.lineplot(x="date", y="hosp_patients", data=covid_data_filtered[covid_data_filtered["continent"] == "Africa"],label='africa')
sns.lineplot(x="date", y="hosp_patients", data=covid_data_filtered[covid_data_filtered["continent"] == "North America"],label='north america')
sns.lineplot(x="date", y="hosp_patients", data=covid_data_filtered[covid_data_filtered["continent"] == "South America"],label='south america')
sns.lineplot(x="date", y="hosp_patients", data=covid_data_filtered[covid_data_filtered["continent"] == "Europe"],label='europe')
sns.lineplot(x="date", y="hosp_patients", data=covid_data_filtered[covid_data_filtered["continent"] == "Oceania"],label='oceania')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('hosp_patients per continent')
plt.show()
```

델타변이바이러스의 특징

전파력 1.6배·입원율 2.26배 높아

```python
plt.figure(figsize=(12, 10))
sns.lineplot(x="date", y="positive_rate", data=covid_data_filtered[covid_data_filtered["continent"] == "Asia"], label='asia')
sns.lineplot(x="date", y="positive_rate", data=covid_data_filtered[covid_data_filtered["continent"] == "Africa"],label='africa')
sns.lineplot(x="date", y="positive_rate", data=covid_data_filtered[covid_data_filtered["continent"] == "North America"],label='north america')
sns.lineplot(x="date", y="positive_rate", data=covid_data_filtered[covid_data_filtered["continent"] == "South America"],label='south america')
sns.lineplot(x="date", y="positive_rate", data=covid_data_filtered[covid_data_filtered["continent"] == "Europe"],label='europe')
sns.lineplot(x="date", y="positive_rate", data=covid_data_filtered[covid_data_filtered["continent"] == "Oceania"],label='oceania')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('positive_rate per continent')
plt.show()
```

세계적으로 코로나19 바이러스가 진정되는 분위기지만 남미에선 얘기가 다르다. 남미는 세계 인구 중 5%밖에 차지하지 않지만 하루 코로나19 사망자 중 25%가 남미에서 쏟아지고 있다고 월스트리트저널(WSJ)이 21일(현지시간) 보도했다.

여기에는 낮은 코로나19 백신 접종률, 전염력이 높은 변이 바이러스의 확산, 도시의 높은 인구 밀집도, 부실한 헬스케어 시스템, 아시아나 아프리카에 비해 훨씬 높은 비만율, 정부의 바이러스 통제 포기 등이 복합적으로 작용한 것으로 풀이된다. 데니스 개럿 전염병 전문가는 "코로나 팬데믹이 미국을 비롯해 전 세계를 휩쓸었지만 남미는 팬데믹의 새 진앙지로 떠오른 상황"이라고 진단했다.
https://news.mt.co.kr/mtview.php?no=2021062215563088668
