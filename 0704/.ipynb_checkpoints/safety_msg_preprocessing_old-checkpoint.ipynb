## 찾은 데이터에서 일일 발송량 계산하기

```python
# pandas series
# msg_count = encouraging_test.groupby(encouraging_test["create_date"].dt.date).size()
# pandas dataframe
msg_count = encouraging_test.groupby(encouraging_test["create_date"].dt.date).count()
msg_count.columns = ["amount"]
msg_count.reset_index(inplace=True)
msg_count["create_date"] = pd.to_datetime(msg_count["create_date"])
msg_count.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20, 10))
# fig, ax = plt.subplots(ncols=1, nrows=2, figsize=(20, 20))
sns.lineplot(data=msg_count, x="create_date", y="amount")
```

```python
from scipy.signal import find_peaks_cwt
import numpy as np

peaks = find_peaks_cwt(msg_count.rolling(7).sum()["amount"].to_numpy(), widths=np.ones(msg_count.rolling(7).sum()["amount"].to_numpy().shape)*2)
print(peaks)
plt.figure(figsize=(20, 10))
# plt.plot(peaks, msg_count.rolling(7).sum()[peaks], "x")
# sns.lineplot(data=msg_count.rolling(7).sum())
plt.plot(msg_count.rolling(7).sum())
plt.plot(peaks, msg_count.rolling(7).sum().iloc[peaks], "x", markersize="10", color="red")
```

```python
msg_count.rolling(7).mean()
```

```python
msg_count_smoothed = msg_count.rolling(7, min_periods=0).sum()
msg_count_smoothed.index = msg_count.create_date
msg_count_smoothed_peaks = msg_count_smoothed.iloc[peaks].index
msg_count_smoothed
```

```python
# plt.figure(figsize=(20, 10))
# plt.plot(msg_count_smoothed)
# plt.plot(peaks, msg_count_smoothed[msg_count_smoothed_peaks], "x", markersize="10", color="red")
# msg_count_smoothed_peaks
msg_count_smoothed.loc[msg_count_smoothed_peaks]
```

```python
peaks = find_peaks_cwt(msg_count.rolling(7).sum()["amount"].to_numpy(), widths=np.ones(msg_count.rolling(7).sum()["amount"].to_numpy().shape)*2)
print(peaks)
plt.figure(figsize=(20, 10))
# plt.plot(peaks, msg_count.rolling(7).sum()[peaks], "x")
# sns.lineplot(data=msg_count.rolling(7).sum())
plt.plot(msg_count_smoothed / 7)
plt.plot((msg_count_smoothed/7).loc[msg_count_smoothed_peaks], "x", markersize="10", color="red")
```
