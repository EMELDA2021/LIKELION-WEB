```python
train.loc[train.Embarked=='C', 'Embarked_num'] = 1
train.loc[train.Embarked=='Q', 'Embarked_num'] = 2
train.loc[train.Embarked=='S', 'Embarked_num'] = 3
```

```python
# Embarked 살펴보기
count = train.Embarked_num.value_counts()
print(count, type(count), count.values, type(count.values))
print(train.Embarked.unique())
list_em = np.array( ['S', 'C', 'Q'] )
```

```python
plt.bar(list_em, count.values, color='red', alpha=0.5)
plt.xlabel("Embarked")
plt.ylabel("Count")
plt.title("Embarked Bar plot")
```

```python
#barplot 그리기
train_sur_1 = train[train.Survived==1] # 생존
train_sur_0 = train[train.Survived==0] # 사망
```

```python
# Embarked 살펴보기
count_1 = train_sur_1.Embarked_num.value_counts()
count_0 = train_sur_0.Embarked_num.value_counts()
print(count_1.values, count_0.values)
list_em = np.array( ['S', 'C', 'Q'] )
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.bar(list_em, count_1.values, color='red', alpha=0.5)
plt.xlabel("Embarked (Survived : 1)", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.ylim([0,500])
plt.title("Embarked Bar plot", fontsize=15)

plt.subplot(1,2,2)
plt.bar(list_em, count_0.values, color='blue', alpha=0.5)
plt.xlabel("Embarked (Survived : 0)", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.ylim([0,500])
plt.title("Embarked Bar plot", fontsize=15)
plt.suptitle("Embarked bar plot")
```

<예준님 자료>

```python
## 승객들 요금 순 정렬 (상위 30명)
x = train.sort_values(by='Fare', ascending=False).head(30)
y = np.array(x['Age'].sort_values())   # 나이 분포
y1 = np.array(x['SibSp'])   # 배우자, 형제 자매 분포
y2 = np.array(x['Parch'])   # 아들/딸 , 부모 분포
y3 = np.array(x['Survived'])   # 생존자 분포
fig = plt.figure(figsize=(8, 4))
plt.hist(y)
fig1 = plt.figure(figsize=(6,2))
plt.hist(y1)
plt.grid()
fig2 = plt.figure(figsize=(6,2))
plt.hist(y2)
plt.grid()
fig3 = plt.figure(figsize=(4,2))
plt.hist(y3)
plt.grid()
plt.show()
```

```python
## 승객들 요금 순 정렬 (하위 30명)
x = train.sort_values(by='Fare', ascending=False).tail(30)
y = np.array(x['Age'].sort_values())   # 나이 분포
y1 = np.array(x['SibSp'])   # 배우자, 형제 자매 분포
y2 = np.array(x['Parch'])   # 아들/딸 , 부모 분포
y3 = np.array(x['Survived'])   # 생존자 분포
fig = plt.figure(figsize=(8, 4))
plt.hist(y)
fig1 = plt.figure(figsize=(6,2))
plt.hist(y1)
plt.grid()
fig2 = plt.figure(figsize=(6,2))
plt.hist(y2)
plt.grid()
fig3 = plt.figure(figsize=(4,2))
plt.hist(y3)
plt.grid()
plt.show()
```

```python
## 모든 승객 Pclass 비율 출력
x = train.sort_values(by='Pclass')
x = x['Pclass'].value_counts()
x = np.array(x)
fig = plt.figure(figsize=(4,4))
plt.pie(x)
plt.legend(['3', '1', '2'])
plt.show()
```

 - 많은 요금을 낸 승객들이 적은 요금을 낸 승객들에 비해 영아와 나이많은 사람이 조금 더 많다.
 - 많은 요금을 낸 승객들에 비해 적은 요금을 낸 승객들이 훨씬 큰 비율로 사망자가 발생했다.
 - 많은 요금을 낸 승객들에 비해 적은 요금을 낸 승객들은 대부분 혼자서 왔다.
